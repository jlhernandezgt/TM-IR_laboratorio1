# Problema 2: 
En el .zip del laboratorio se le compartió un conjunto de imágenes en escala de grises (imagen1, imagen2, perro) estas imágenes fueron creadas utilizando una escala de grises en 3D, cree una función que dadas las 3 imágenes se construya la imagen original a color.


```python
def fn_problema2(imagen_gr, imagen_gv, imagen_ga):
    imgn_gr = convert2(imagen_gr)
    imgn_gv = convert2(imagen_gv)
    imgn_ga = convert2(imagen_ga)
    alto = imgn_gr.shape[0]
    ancho = imgn_gr.shape[1]
    full_color = np.zeros((alto, ancho, 3))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel_gr = imgn_gr[i, j]
            pixel_gv = imgn_gv[i, j]
            pixel_ga = imgn_ga[i, j]                      
            colores = tuple((pixel_ga[0],pixel_gv[1], pixel_gr[2]))                            
            full_color[i, j] = colores
    cv2.imwrite("output_full_color.jpg", full_color)
    output_color = cv2.imread("output_full_color.jpg")
    convert(output_color)    
```

```python
def fn_problema3(imagen):
    imgn = convert2(imagen)
    alto = imgn.shape[0]
    ancho = imgn.shape[1]
    G1 = np.zeros((alto, ancho, 1))
    G2 = np.zeros((alto, ancho, 1))
    G3 = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imgn[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            G1[i, j] = int(0.359*blue + 0.487*green + 0.684*red)
            G2[i, j] = int(0.299*blue + 0.587*green + 0.383*red)
            G3[i, j] = int(0.119*blue + 0.257*green + 0.444*red)
    cv2.imwrite("gris_1.jpg", G1)
    cv2.imwrite("gris_2.jpg", G2)
    cv2.imwrite("gris_3.jpg", G3)
    outputG1 = cv2.imread("gris_1.jpg")
    outputG2 = cv2.imread("gris_2.jpg")
    outputG3 = cv2.imread("gris_3.jpg")
    outputG1 = cv2.cvtColor(outputG1, cv2.COLOR_BGR2RGB)
    outputG2 = cv2.cvtColor(outputG2, cv2.COLOR_BGR2RGB)
    outputG3 = cv2.cvtColor(outputG3, cv2.COLOR_BGR2RGB)
    convert(outputG1)
    convert(outputG2)
    convert(outputG3)
```

# Problema 4:
Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.


```python
problema4 = cv2.imread("problema4.jpg")
```

```python
def fn_problema4(imagen):
    imgn = convert2(imagen)
    alto = imgn.shape[0]
    ancho = imgn.shape[1]
    GScale = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imgn[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            GScale[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
    cv2.imwrite("gris_problema4.jpg", GScale)
    outputGScale = cv2.imread("gris_problema4.jpg")
    outputGScale = cv2.cvtColor(outputGScale, cv2.COLOR_BGR2RGB)
    
    vertical_original = imgn.mean()
    vertical_gris = outputGScale.mean()
    
    plt.imshow(imgn)
    plt.title('Imagen Original')
    plt.show()
    
    color = ('b','g','r')
    for i, c in enumerate(color):
        hist = cv2.calcHist([imgn], [i], None, [256], [0, 256])
        plt.plot(hist, color = c)
        plt.xlim([0,256])
        plt.title('Histograma Color')
        plt.axvline(x=vertical_original,color=  'red', ymin=0.0, ymax=0.9)
    plt.show()    
    
    plt.imshow(outputGScale)
    plt.title('Escala de Grises')
    plt.show()
    
    plt.hist(outputGScale.ravel(),256,[0,256])
    plt.title('Histograma Escala Grises')
    plt.axvline(x=vertical_gris,color=  'red', ymin=0.0, ymax=0.9)
    plt.show()
    
```

```python
fn_problema4(problema4)
```

Investigue en que consiste el enfoque de escala de grises ponderado, luego de esto implemente una función que dada una imagen, realice una escala de grises ponderada (notar que no existe una solución única).


## Escala de Grises

Es el equivalente a la luminancia de la imagen, el ojo percibe distintas intesidades de luz en función del color que se observe, esto es debido a la respuesta del ojo al espectro visible la cual se puede observar en la figura, por esa razón el cálculo del equivalente blanco y negro (escala de grises o luminancia) de la imagen debe realizarse como una media ponderada de las distintas componentes de color de cada pixel.  La expresión matemática de ese fenómeno y los factores de ponderación de cada componente de color nos indican la sensibilidad del ojo humano a las frecuencias del espectro cercanas al rojo, verde y azul.  

Por tanto, para realizar esta conversión llega con aplicar esta ecuación a cada pixel de la imagen true-color. Aparecería una nueva matriz de un byte por pixel que daría la información de luminancia.

Según la importancia y otros indicadores, los tres componentes se ponderan y se promedian con diferentes ponderaciones. Dado que el ojo humano es el más sensible al verde y el menos sensible al azul, se puede obtener una imagen en escala de grises más razonable ponderando los tres componentes RGB de la siguiente manera.
#### Y = R*0.3+G*0.59+B*0.11




```python
def fn_problema5(imagen):
    imgn = convert2(imagen)
    alto = imgn.shape[0]
    ancho = imgn.shape[1]
    G1 = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = imgn[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            G1[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
    cv2.imwrite("gris_problema5.jpg", G1)
    outputG1 = cv2.imread("gris_problema5.jpg")
    outputG1 = cv2.cvtColor(outputG1, cv2.COLOR_BGR2RGB)
    convert(outputG1)
```

## Modelo de color HSV

El modelo HSV fue definido en 1978 por Alvy Ray Smith y es la representación tridimensional del color basado en los componentes de tinte, matiz o tonalidad, saturación  y brillo o valor, y estos valores suelen representarse en un diagrama circular.

A diferencia del modelo RGB ampliamente usado en los monitores, televisores, etc., el color HSV sigue una representación más parecida a las coordenadas cilíndricas. Además es una representación más cercana a la forma en que los humanos perciben los colores y sus propiedades, pues se agrupan las tonalidades de color, lo cual es distinto al caso RGB donde los colores no están necesariamente tan agrupados.

Al definirse en 1978, en pleno auge de la televisión en colores, por Alvy Ray Smith, se intentaba, a diferencia de los modelos RGB, YUV, YIQ, representar colores desde un punto de vista más próximo a la percepción humana. Si bien los otros modelos mencionados asistían eficientemente el problema tecnológico de la representación y trasmisión de imágenes en colores, las asociaciones entre colores que comúnmente suceden en la visión humana, quedaban fuera de los anteriores espacios de color. HSV viene a ser por tanto una importante herramienta para el humano que interactúa con las tecnologías de representación de color.

### Espacios de color HSV

Los artistas suelen usar espacios de color de matiz, saturación y valor (HSV). "Hue" es lo que normalmente pensamos como color. Es el atributo de un color por el que le asignamos un nombre como "rojo" o "azul". "Value" es otra palabra para "ligereza", el atributo de un color que hace que parezca equivalente a algún tono de gris entre blanco y negro. La saturación es una medida de lo diferente que aparece un color de un gris de la misma ligereza. La saturación cero indica ningún matiz, solo escala gris. El espacio de color HSV se normaliza.



En la ilustración anterior se muestra un dibujo de líneas de espacio HSV en forma de hexcono. Cada una de sus secciones transversales es un hexágono. En los vértices de cada sección transversal se encuentran los colores rojo, amarillo, verde, cian, azul y magenta. Para especificar un color en el espacio HSV, se indica un ángulo de matiz, el nivel de cromática y el nivel de ligereza. Un ángulo de matiz de cero es rojo. El ángulo de matiz aumenta en una dirección en sentido contrario a las agujas del reloj. Los colores complementarios están separados por 180.

Los espacios de color de HSV pueden ser dependientes del dispositivo o independientes del dispositivo.
